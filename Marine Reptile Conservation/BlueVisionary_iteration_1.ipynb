# Finding Conservation Status

```python
def fetch_species_data_name(species_name, token):
    """
    Function to fetch the conservation status for each species from the list create above
    """
    
    api_url =f"https://apiv3.iucnredlist.org/api/v3/species/{species_name}?token={token}"

    try:
        response = requests.get(api_url, timeout=10)
        response.raise_for_status()
        species_data = response.json()
        if 'result' in species_data and species_data['result']:
            result = species_data['result'][0]
        # Filter results based on the scientific names in the DataFrame
        # Return the relevant fields
            return {
                'scientificName': result.get('scientific_name',species_name),
                'category': result.get('category')
            }
        else:
            print(f"No result found for {species_name}")
            return None
        
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
        
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Connection error occurred: {conn_err}")
        
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout error occurred: {timeout_err}")
        
    except requests.exceptions.RequestException as req_err:
        print(f"An error occurred: {req_err}")
        
    return None
```

```python
# List to store fetched conservation data
fetched_data_final_status = []

# Loop through each scientific name list
for species_name in scientific_names:
    species_data = fetch_species_data_name(species_name, token)
    if species_data:
        fetched_data_final_status.append(species_data)
    else:
        # In case no data found, append a row with None
        fetched_data_final_status.append({
            'scientificName': species_name,
            'category': None
        })

# Creating the dataframe from the fetched data
fetched_species_df = pd.DataFrame(fetched_data_final_status )
```

# Finding Narrative Text 

```python
def clean_html(raw_html):
    """
    Function for removing HTML tags from a string and decoding HTML entities.
    And to handle cases where the input is not a string.
    """
    if isinstance(raw_html, str):
        # Remove HTML tags
        clean_text = re.sub(r'<.*?>', '', raw_html)
         # Decode HTML entities
        clean_text = html.unescape(clean_text) 
        return clean_text
    return raw_html 
```

```python

def fetch_species_narrative(species_name, token):
    """
    Function to find out the narrative about threats and habitats to each species    
    """
    api_url =f"https://apiv3.iucnredlist.org/api/v3/species/narrative/{species_name}?token={token}"
    try:
         response = requests.get(api_url, timeout=10)
         response.raise_for_status()
        
         species_data = response.json()
         if 'result' in species_data and species_data['result']:
             result = species_data['result'][0]
             
             # Clean the habitat field
             raw_habitat = result.get('habitat', '')
             clean_habitat = clean_html(raw_habitat)

             raw_threats = result.get('threats', '')
             clean_threats = clean_html(raw_threats)
             
             # Filter results based on the scientific names in the DataFrame
             
        # Return the relevant fields
             return {
                 'scientificName': result.get('name',species_name),
                 'populationtrend': result.get('populationtrend'), 
                 'habitat': clean_habitat,
                 'threats': clean_threats
             }
             
         else:
             print(f"No result found for {species_name}")
             return None
            
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
        
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Connection error occurred: {conn_err}")
        
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout error occurred: {timeout_err}")
        
    except requests.exceptions.RequestException as req_err:
        print(f"An error occurred: {req_err}")
    return []

```

```python
# List to store fetched data
species_narratives = []

# Loop through each scientific name in your DataFrame
for each_name in scientific_names:
    species_data_narrative = fetch_species_narrative(each_name, token)
    if species_data_narrative:
        species_narratives.append(species_data_narrative)
    #else:
        # If no data found or error, append a row with None or empty values
        fetched_data_final_status.append({
            'scientificName': each_name,
            'category': None
        })

# Convert the fetched data list to a DataFrame
fetched_species_df_narrative = pd.DataFrame(species_narratives)
```

# STATE BOUNDARY GEOJSON

```python
# Opening geojson state file
with open('georef-australia-state@public.geojson', 'r') as f:
    state_data = json.load(f)
```

```python
# Check the columns (keys) in the properties of the first feature
feature = state_data['features'][0]
columns = feature['properties'].keys()

print("Columns in the GeoJSON file:")
for column in columns:
    print(column)
```

```python
# Important Columns
imp_cols = ["ste_name"]
```

```python
# creating an empty list to store filtered data
filtered_states = []
for feature in state_data['features']:
    properties = feature['properties']
    if properties['ste_name'] != ["Other Territories"]:
        filtered_feature = {
            "type": "Feature",
            "properties": {key: properties[key][0] for key in imp_cols},
            "geometry": feature["geometry"]
        }
        filtered_states.append(filtered_feature)
```

```python
# Creating the GeoJSON structure from filtered data
cleaned_states = {
    "type": "FeatureCollection",
    "features": filtered_states 
}

# Save the filtered data to a new GeoJSON file
with open('FINAL_STATE_data_new.json', 'w') as f:
    json.dump(cleaned_states, f, indent=2)
```
